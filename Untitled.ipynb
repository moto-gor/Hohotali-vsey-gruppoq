```python
weather = pd.read_csv('C:/Users/Ya/anac/data/weather.csv')
```

```python
weather.isna().sum()
```

```python
weather.info()
```

```python
for col in ['precipitation', 'snow fall', 'snow depth']:
    print(weather[col].unique())
    print('-' * 10)
```

```python
for col in ['precipitation', 'snow fall', 'snow depth']:
    weather.loc[weather[col] == 'T', col] = 0
    weather[col] = weather[col].astype(np.float64)
```

```python
weather.describe()
```

```python
for temp in ['maximum temperature', 'minimum temperature','average temperature']:
    weather[temp] = weather[temp].apply(lambda x: (x-32)/1.8)
```

```python
weather.describe()
```

```python
mainOnmainOn['date'] = mainOn['pickup_datetime'].apply(lambda x: str(x.day) + '-' + str(x.month) + '-' + str(x.year))
mainOn = mainOn.merge(weather, on='date').drop(['date'], axis=1)
mainOn
```

```python
mainOn['year'] = mainOn['dropoff_datetime'].apply(lambda x: x.year)
mainOn['month'] = mainOn['dropoff_datetime'].apply(lambda x: x.month)
mainOn['day'] = mainOn['dropoff_datetime'].apply(lambda x: x.day)
mainOn['hour'] = mainOn['dropoff_datetime'].apply(lambda x: x.hour)
mainOn
```

```python
mainOn.drop('dropoff_datetime', axis=1, inplace=True)
```

```python
mainOn['category_travel_time']=mainOn['trip_duration']
mainOn.info()
```

```python
q1=mainOn.category_travel_time.quantile(0.25)
q2=mainOn.category_travel_time.quantile(0.5)
q3=mainOn.category_travel_time.quantile(0.75)

mainOn.category_travel_time=np.where((mainOn.category_travel_time<=q1),1,mainOn.category_travel_time)
mainOn.category_travel_time=np.where(((mainOn.category_travel_time<q2)&(mainOn.category_travel_time>q1)),2,mainOn.category_travel_time)
mainOn.category_travel_time=np.where(((mainOn.category_travel_time<q3)&(mainOn.category_travel_time>=q2)),3,mainOn.category_travel_time)
mainOn.category_travel_time=np.where((mainOn.category_travel_time>=q3),4,mainOn.category_travel_time)

mainOn.drop('trip_duration', axis=1, inplace=True)
mainOn.info()
```

```python
mainOn.to_csv('C1_result.csv', index=False)
```
